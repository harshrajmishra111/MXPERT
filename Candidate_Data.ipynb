```python
import pandas as pd

# Load the first dataset
dataset1_path = ""
dataset1 = pd.read_csv(dataset1_path)

# Select the required columns from the first dataset
selected_columns1 = ['EdLevel', 'YearsCodePro', 'HaveWorkedWith', 'ComputerSkills']
dataset1_selected = dataset1[selected_columns1]

# Load the second dataset

dataset2_path = ""
dataset2 = pd.read_csv(dataset2_path)

# Select the required column from the second dataset
selected_columns2 = ['cand_name']
dataset2_selected = dataset2[selected_columns2]

# Combine both datasets into one DataFrame column-wise
combined_df = pd.concat([dataset2_selected, dataset1_selected], axis=1)

# Add an ID column
combined_df.insert(0, 'ID', range(1, len(combined_df) + 1))

# Reorder columns so 'cand_name' is after 'ID'
combined_df = combined_df[['ID', 'cand_name', 'EdLevel', 'YearsCodePro', 'HaveWorkedWith', 'ComputerSkills']]

# Rename columns to the desired names
combined_df.columns = ['ID', 'Candidate Name', 'Education', 'Experience', 'Skills', 'Number of Skills']

# Display the final DataFrame
print(combined_df.head())
```

```python
import pandas as pd

# Print the summary of the DataFrame
print("Summary of the DataFrame:")
print(combined_df.describe(include='all'))

# Check for missing values

print("\nMissing values in each column:")
print(combined_df.isnull().sum())
```
