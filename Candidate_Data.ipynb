```python
import pandas as pd

# Load the first dataset
dataset1_path = ""
dataset1 = pd.read_csv(dataset1_path)

# Select the required columns from the first dataset
selected_columns1 = ['EdLevel', 'YearsCodePro', 'HaveWorkedWith', 'ComputerSkills']
dataset1_selected = dataset1[selected_columns1]

# Load the second dataset

dataset2_path = ""
dataset2 = pd.read_csv(dataset2_path)

# Select the required column from the second dataset
selected_columns2 = ['cand_name']
dataset2_selected = dataset2[selected_columns2]

# Combine both datasets into one DataFrame column-wise
combined_df = pd.concat([dataset2_selected, dataset1_selected], axis=1)

# Add an ID column
combined_df.insert(0, 'ID', range(1, len(combined_df) + 1))

# Reorder columns so 'cand_name' is after 'ID'
combined_df = combined_df[['ID', 'cand_name', 'EdLevel', 'YearsCodePro', 'HaveWorkedWith', 'ComputerSkills']]

# Rename columns to the desired names
combined_df.columns = ['ID', 'Candidate Name', 'Education', 'Experience', 'Skills', 'Number of Skills']

# Display the final DataFrame
print(combined_df.head())
```

```python
import pandas as pd

# Print the summary of the DataFrame
print("Summary of the DataFrame:")
print(combined_df.describe(include='all'))

# Check for missing values

print("\nMissing values in each column:")
print(combined_df.isnull().sum())
```

```python
import pandas as pd

# Remove duplicates based on 'Candidate Name', keeping the first occurrence
unique_df = combined_df.drop_duplicates(subset=['Candidate Name'], keep='first')

# Check the size of the unique dataset after removing duplicate candidate names
print(f"Number of unique records based on Candidate Name: {unique_df.shape[0]}")

# If the dataset still has more than 50,000 records, reduce it to 50,000
if unique_df.shape[0] > 50000:
    reduced_df = unique_df.sample(n=50000, random_state=42)
else:
    reduced_df = unique_df
    
# Reset index for the reduced dataset
reduced_df.reset_index(drop=True, inplace=True)

# Check the size of the reduced dataset
print(f"Final dataset size after removing duplicates by Candidate Name: {reduced_df.shape[0]}")

# Display the summary of the reduced dataset
print(reduced_df.describe(include='all'))

# Check for missing values in the reduced dataset
print("\nMissing values in each column:")
print(reduced_df.isnull().sum())
```

```python
import pandas as pd

# Handle missing 'Candidate Name'
# Drop the row with missing Candidate Name

reduced_df = reduced_df.dropna(subset=['Candidate Name'])

# Handle missing 'Skills'
# Drop rows where Skills are missing
reduced_df = reduced_df.dropna(subset=['Skills'])

# Reset index after dropping rows
reduced_df.reset_index(drop=True, inplace=True)

# Check for any remaining missing values
print("\nRemaining missing values in each column:")
print(reduced_df.isnull().sum())

# Display summary of the cleaned dataset
print("\nSummary of the cleaned DataFrame:")
print(reduced_df.describe(include='all'))
```

```python
import pandas as pd

# Function to sum digits and return a single digit capped at 7
def sum_to_single_digit(n):
    while n >= 10:
        n = sum(int(digit) for digit in str(n))
    return min(n, 7)  # Cap at 7

# Apply the function to the Experience column for the entire dataset
reduced_df['Experience'] = reduced_df['Experience'].apply(sum_to_single_digit)

# Display the updated DataFrame summary
print("\nSummary of the cleaned DataFrame with modified Experience:")
print(reduced_df.describe(include='all'))
```
